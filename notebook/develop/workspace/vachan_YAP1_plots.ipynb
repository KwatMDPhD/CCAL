# Plotting YAP1 to CYRG61 With Bokeh

```python
#Created with Python 3

import sys
sys.path.insert(0, '../../')
import ccal
#from bokeh.charts import Scatter, output_file, show
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

#read gct files as pandas dataframe
proteins = ccal.support.read_gct('C:/Users/GravityGoose/Documents/MyStuff/Cancer Center/HNSCC_datasets/HNSCC_RPPA.v1.gct')
gene_exp = ccal.support.read_gct('C:/Users/GravityGoose/Documents/MyStuff/Cancer Center/HNSCC_datasets/HNSCC_EXP.v1.gct')
mut_cna = ccal.support.read_gct('C:/Users/GravityGoose/Documents/MyStuff/Cancer Center/HNSCC_datasets/HNSCC_MUT_CNA_plus_cbio_FAT1.alterations.v1.3.gct')

print("\n GAGGFAGFAD \n")

prot_samp_names = list(proteins.columns.values)
gene_samp_names = list(gene_exp.columns.values)
mut_cna_samp_names = list(mut_cna.columns.values)

#intersection between gene and protein datasets
intersec = []
for ps in mut_cna:
    for gs in gene_samp_names:
        if ps == gs:
            intersec.append(ps)
            
#Extract the row of specified gene         
cyr61 = gene_exp.loc['CYR61']
yap1 = gene_exp.loc['YAP1']
ctgf = gene_exp.loc['CTGF']

#Bokeh plot the genes

#Plot YAP1 vs CYR61
output_notebook() #output to jupyter instead of html file
p1 = figure(plot_width=800, plot_height=500, title="CYR61 vs YAP1")
p1.circle(cyr61, yap1, size=10, color="navy", alpha=0.5)
p1.xaxis.axis_label = "CYR61"
p1.yaxis.axis_label = "YAP1"
show(p1)

#plot YAP1 vs CTGF
p2 = figure(plot_width = 800, plot_height=500, title = "CTGF vs YAP1")
p2.circle(ctgf, yap1, size=10, color="red", alpha=0.4)
p2.xaxis.axis_label = "CTGF"
p2.yaxis.axis_label = "YAP1"
show(p2)
```
