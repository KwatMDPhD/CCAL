```python
import cProfile
#import timeit
import time
from library.ccba import *
from scipy.stats import pearsonr
import seaborn as sns
%matplotlib inline
```

```python
def runtime(function, n_range, verbose=False):
    ns = []
    runtimes = []
    for i in n_range:
        n = 10 * (i + 1)
        if verbose:
            print(n)
        x = np.random.rand(n)
        y = np.random.rand(n)
        t0 = time.time()

        function(x, y)

        runtime = time.time() - t0
        ns.append(n)
        runtimes.append(runtime)
        #cProfile.run(func)
    return ns, runtimes
```

```python
ns, runtimes = runtime(information_coefficient, n_range=list(range(1000)))
ax = sns.pointplot(x=ns, y=runtimes)
```

```python
ns, runtimes = runtime(pearsonr, n_range=list(range(1000)), verbose=False)
ax = sns.pointplot(x=ns, y=runtimes)
```
